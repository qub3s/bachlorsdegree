```python
import numpy as np
import matplotlib.pyplot as plt

with open("logs/own_approach_first_train", "r") as my_file:
    file_content = my_file.read()

file_content = file_content.split("\n")


keyword = [ "iou_error", "lam_dis", "lam_col", "lam_cen", "number_of_masks", "bb_max_size", "threshhold_depth" ]

para_trends = [[], [], [], [], [], [], []]

round_nr = "0"

for line in file_content:
    if "Round" in line and round_nr in line:
        # 
    for i,key in enumerate(keyword):
        if key in line:
            temp = str(line)
            temp = temp.replace(key, "")
            temp = temp.replace(" ", "")
            temp = temp.replace(":", "")
            para_trends[i].append(float(temp))



# don't add the round if nothing could have changed anyway
```
